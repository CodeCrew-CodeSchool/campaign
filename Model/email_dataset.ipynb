```python
import pandas as pd
import random

filename = 'email_dataset.csv'
df = pd.read_csv(filename)

# Clean column headers
df.columns = df.columns.str.strip()

# Check dataset structure
print("Columns in dataset:", df.columns)
print("Dataset info:")
print(df.info())

# Preview first few rows
print("First few rows:")
print(df.head())

# Example: Filter templates with keyword in 'Subject'
keyword = 'Sale'  # Change as needed
filtered_df = df[df['Subject'].str.contains(keyword, case=False, na=False)]
print(f"\nTemplates with '{keyword}':")
print(filtered_df)

# Sample multiple templates
print("\nSample 3 random email templates:")
for _ in range(3):
    sample = df.sample(1).iloc[0]
    subject = sample['Subject']
    description = sample['Description']
    body = sample['Body']
    closing = sample['Closing']
    email_content = (
        f"Subject: {subject}\n"
        f"Description: {description}\n\n"
        f"{body}\n\n"
        f"{closing}"
    )
    print("-" * 50)
    print(email_content)

# Generate a full email template
def generate_full_email(dataframe):
    sample = dataframe.sample(1).iloc[0]
    subject = sample['Subject']
    description = sample['Description']
    body = sample['Body']
    closing = sample['Closing']
    email_content = (
        f"Subject: {subject}\n"
        f"Description: {description}\n\n"
        f"{body}\n\n"
        f"{closing}"
    )
    return email_content

# Generate and print 3 sample emails
print("\nGenerated sample emails with all columns:")
for _ in range(3):
    email = generate_full_email(df)
    print("="*50)
    print(email)

# Save a subset for testing
subset_df = df.sample(10)
subset_df.to_csv('subset_emails.csv', index=False)
print("\nSaved a subset of 10 templates to 'subset_emails.csv' for further testing.")
```
